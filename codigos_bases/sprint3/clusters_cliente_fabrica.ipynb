```python
from sklearn.cluster import KMeans
```

```python
X = df_concat[['LAT', 'LONG']]
```

```python
# Executar o algoritmo K-means com 7 clusters
kmeans = KMeans(n_clusters=7, random_state=42)
```

```python
X = df_concat[['LAT', 'LONG']]
```

```python
# Obter os rótulos de cluster para cada ponto de dados
labels = kmeans.labels_

# Obter as coordenadas dos centróides
centroids = kmeans.cluster_centers_
```

```python
pred_cluster = kmeans.fit_predict(X)
# Criar um DataFrame com os clusters e centróides
df_clusters = pd.DataFrame({'CODIGO': df_concat['CODIGO'],'NO_MUN': df_concat['NO_MUN'],'LAT': df_concat['LAT'], 'LONG': df_concat['LONG'], 'cluster': labels})
df_centroids = pd.DataFrame({'LAT_CT': centroids[:, 0], 'LONG_CT': centroids[:, 1], 'cluster': range(len(centroids))})
```

```python
#incluindo a base com as quantidades anuais de cada cliente
base_rota_qtd=r"C:\Users\jetin\Python_Códigos\Roteirização\rotas_agrup_vd_qtd.csv"

dict_type_rota_qtd={'cod_cliente':str}
```

```python
df_rota_qtd = pd.read_csv(base_rota_qtd, sep=",", dtype=dict_type_rota_qtd)
```

```python
# enriquecimento da base clusters com a qtd anual transportada por cliente
df_merge_2=pd.merge(df_merge, df_rota_qtd, on=['CODIGO', 'LAT', 'LONG'], how='outer')
```

```python
df_merge_2 = df_merge_2.fillna(0)
```

```python
df_merge_2.head()
```

```python
#Agrupamento dos cluster, e soma da qtda_transportada em cada um
chave_cluster_qtd=['cluster','LAT_CT','LONG_CT']
var_int_cluster_qtd=['qtd_transportada']

df_clusters_qtd_agrup=df_merge_2.groupby(chave_cluster_qtd)[var_int_cluster_qtd].sum().reset_index()
```

```python
# base final para aplicação do COG
df_clusters_qtd_agrup
```

#### Aplicando COG para encontrar o CD 

```python
demanda_total = df_clusters_qtd_agrup['qtd_transportada'].sum()
```

```python
#Criando o TUPLE de coordenadas para cálculo do COG
df_clusters_qtd_agrup['coordenadas']= df_clusters_qtd_agrup.apply(lambda row: (row['LAT_CT'], row['LONG_CT']), axis=1)
```

```python
#Cálculo da coordenada X do CD usando COG
df_clusters_qtd_agrup['x_weighted'] = df_clusters_qtd_agrup.apply(lambda row: row['qtd_transportada'] * row['coordenadas'][0], axis=1)
x_cog = df_clusters_qtd_agrup['x_weighted'].sum() / demanda_total
```

```python
#Cálculo da coordenada Y do CD usando COG
df_clusters_qtd_agrup['y_weighted'] = df_clusters_qtd_agrup.apply(lambda row: row['qtd_transportada'] * row['coordenadas'][1], axis=1)
y_cog = df_clusters_qtd_agrup['y_weighted'].sum() / demanda_total
```

```python
#Resultado da coordenada Lat, Long do CD
print('O centro de distribuição deve ser alocado próximo das coordenadas: {},{}'.format(x_cog, y_cog))
```
